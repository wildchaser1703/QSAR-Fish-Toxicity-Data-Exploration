```python
#Annot=True puts the data in the boxes respectively
sns.heatmap(data.corr(),annot=True)
```

It looks like the Target-LC50 variable shares a positive correlation with MLOGP and RDCHI whereas it shares a moderately negative correlation with GATS1p.

```python
# Plotting individual scatter plots to further analyse the linearity
```

```python
data.plot(kind="scatter", x = "MLOGP" , y = "Target- LC50 [-LOG(mol/L)]")
```

As the value of x goes on increasing from -6 to 8 the value of y keeps on increasing from 0 to 10.Therefore,the two variables share a positive relationship.

```python
data.plot(kind = "scatter", x = "RDCHI", y ="Target- LC50 [-LOG(mol/L)]")
```

As the value of x goes on increasing from 1 to 6 the value of y keeps on increasing from 0 to 10.Therefore,the two variables share a positive correlation relationship. However the points seem to scatter after a certain value. Hence, the correlation value seems to be lower.

```python
data.plot(kind="scatter", x ="GATS1p", y="Target- LC50 [-LOG(mol/L)]" )
```

As the value of x goes on increasing from 0.5 to 2.5 the value of y keeps on decreasing from 10 to 0.Therefore,the two variables share a negative correlation relationship. However the points seem to scatter after a certain value. Hence, the correlation value seems to be lower.

It can be inferred from the following graphs that the target variables share a linear relationship with each other and hence a linear regression model can be used here.
